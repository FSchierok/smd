```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

%matplotlib inline
plt.rcParams.update({'figure.figsize': (12, 8), 'font.size': 12})
plt.style.use(['dark_background'])
np.random.seed(42)
```

## Teil a

```python
F = 5  # Freiheitsgrade
chis = np.random.chisquare(F, 100)
```

## Teil b

```python
fig1, ax1 = plt.subplots()
counts, binEdges = np.histogram(chis, 7)  # powerd by Stackoverflow
binCenters = 0.5*(binEdges[1:]+binEdges[:-1])
ax1.bar(binCenters, counts, yerr=np.sqrt(counts),
        ecolor="tab:brown", width=2, label="Teilaufgabe a")
ax1.legend()
```

## Teil c

```python
fig2, ax2 = plt.subplots()
# Histogram in numpy erstellen, um mit den counts rechnen zu k√∂nnen
counts, binEdges = np.histogram(chis, 7, density=True)
ax2.bar(binCenters, counts, yerr=np.sqrt(counts/len(chis)),
        ecolor="tab:brown", width=2, label="Teilaufgabe a")
x_plot = np.linspace(chi2.ppf(0.01, F), chi2.ppf(0.99, F), 50)
ax2.plot(x_plot, chi2.pdf(x_plot, F), color="tab:orange", label="pdf")
ax2.legend()
```

## Teil d

```python
fit = chi2.fit(chis, F)
ax2.plot(x_plot, chi2.pdf(x_plot, fit[0]), color="tab:green", label="Fit")
ax2.legend()
fig2
```
