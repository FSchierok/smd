## Teilaufgabe a

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams.update({'figure.figsize': (15, 10), 'font.size': 12})
```

```python
def f(x):
    return (x**3+1/3) - (x**3 -1/3)

f_true = 2/3

def Test(fkt, real, test_range):
    print("1% Test:")
    for x in np.logspace(0, test_range, 50):
        if np.abs((fkt(x)-real)/real) >= 0.01:
            print(x) 
    print("\n 0 Test:")
    for x in np.logspace(0, test_range, 50):
        if fkt(x) == 0:
            print(x)
            
Test(f, f_true, 6)
# 1% bei ca 50000
# 0 bei ca 200000
```

## Teilaufgabe b

```python
def g(x):
    return ((3+(x**3) /3)- (3-(x**3) /3))/x**3

g_true = 2/3

Test(g, g_true, 102)
# 1% nicht zu finden
# 0 nicht zu finden
```

## Teilaufgabe c

```python
x_plot = np.logspace(0, 10)
plt.plot(x_plot, f(x_plot), label = "f(x)")
plt.plot(x_plot, g(x_plot), label = "g(x)")
plt.hlines(y = 2/3, xmin = min(x_plot), xmax=max(x_plot),
           label = r"$\dfrac{2}{3}$, wahrer Wert")
plt.xscale("log")
plt.xlabel("x")
plt.legend()
```

## Teilaufgabe d

```python
plt.clf()
fig = plt.figure()
x_32 = np.logspace(0,10, dtype="float32")
x_64 = np.logspace(0,10, dtype="float64")
ax11 = plt.subplot(2, 2, 1)
ax12 = plt.subplot(2, 2, 2)
ax21 = plt.subplot(2, 2, 3)
ax22 = plt.subplot(2, 2, 4)

ax11.plot(x_32, f(x_32), label = "f_32")
ax11.plot(x_32, g(x_32), label = "g_32")
ax11.set_xscale("log")
ax11.set_title("f_32 & g_32")
ax11.legend()


ax12.plot(x_64, f(x_64), label = "f_64")
ax12.plot(x_64, g(x_64), label = "g_64")
ax12.set_xscale("log")
ax12.set_title("f_64 & g_64")
ax12.legend()


ax21.plot(x_32, f(x_32), label = "f_32")
ax21.plot(x_64, f(x_64), label = "f_64")
ax21.set_xscale("log")
ax21.set_title("f_32 & f_64")
ax21.legend()


ax22.plot(x_32, g(x_32), label = "g_32")
ax22.plot(x_64, g(x_64), label = "g_64")
ax22.set_xscale("log")
ax22.set_title("g_32 & g_64")
ax22.legend()
```

Man sieht sehr deutlich das die $x_{64}$ Funktionen ihren Sollwert länger halten. Zudem ist wieder zu erkennen, das $g(x)$ stabiler ist. Dies ist im vierten Plot zu sehen, wo die $y$-Skala in der Größenordnung $10^{-7}$ läuft und erst dadurch Schwankungen sichtbar werden.
